```python
import mysql.connector

# Create a connection to the MySQL server
conn = mysql.connector.connect(user='root', password='root', host='localhost')

# Create a cursor to interact with the MySQL server
cursor = conn.cursor()
```

```python
# Create a new database named 'library'
cursor.execute("CREATE DATABASE IF NOT EXISTS Sysusage")

# Switch to the 'library' database
cursor.execute("USE Sysusage")
cursor.close()
conn.close()
```

```python
import numpy as np
from tensorflow import keras
from tensorflow.keras import layers

# Assuming you have your training data
# X_train = ...
# y_train = ...

# For demonstration purposes, let's create some dummy data
X_train = np.random.rand(100, 4)  # Assuming 100 samples with 4 features each
y_train = np.random.rand(100, 1)   # Assuming 100 samples with 1 target value each

# Define a simple neural network model
model = keras.Sequential([
    layers.Dense(64, activation='relu', input_shape=(4,)),  # Assuming 4 input features
    layers.Dense(32, activation='relu'),
    layers.Dense(1, activation='linear')  # Output layer
])

# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error', metrics=['accuracy'])

# Train the model
model.fit(X_train, y_train, epochs=10, batch_size=32, validation_split=0.2)

# Save the trained model
model.save('usage_model.h5')

```

```python
import mysql.connector
import psutil
import time
from tabulate import tabulate
import csv
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

# Connect to MySQL
mydb = mysql.connector.connect(
    host="localhost",
    user="root",
    password="root",
    database="Sysusage"
)

# Create a cursor object to execute queries
mycursor = mydb.cursor()

# Create table if not exists, adding columns for CPU, Memory, Disk, and I/O usage
mycursor.execute("""
    CREATE TABLE IF NOT EXISTS `Usage` (
        ID INT AUTO_INCREMENT PRIMARY KEY,
        CPU_Usage FLOAT,
        Memory_Usage FLOAT,
        Disk_Usage FLOAT,
        IO_Usage FLOAT
    )
""")
mydb.commit()
print("Table `Usage` is ready in the database.")

# Function to insert CPU, Memory, Disk, and I/O usage into the database
def insert_usage(cpu_usage, memory_usage, disk_usage, io_usage):
    sql = "INSERT INTO `Usage` (CPU_Usage, Memory_Usage, Disk_Usage, IO_Usage) VALUES (%s, %s, %s, %s)"
    val = (cpu_usage, memory_usage, disk_usage, io_usage)
    mycursor.execute(sql, val)
    mydb.commit()

# Function to retrieve CPU, Memory, Disk, and I/O usage from the database
def get_usage():
    mycursor.execute("SELECT * FROM `Usage`")
    return mycursor.fetchall()

# Function to display CPU, Memory, Disk, and I/O usage in a tabular format
def display_usage():
    usage_data = get_usage()
    headers = ["ID", "CPU Usage (%)", "Memory Usage (%)", "Disk Usage (%)", "IO Usage (MB)"]
    print(tabulate(usage_data, headers=headers, tablefmt="grid"))

# Function to write data to a CSV file
def write_to_csv(data, filename):
    with open(filename, mode='w', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(["ID", "CPU Usage (%)", "Memory Usage (%)", "Disk Usage (%)", "IO Usage (MB)"])
        writer.writerows(data)
    print(f"Data written to {filename}")

# Function to calculate weightage using a neural network model
def calculate_weightage(data):
    # Load the trained neural network model
    model = keras.models.load_model('usage_model.h5')

    # Predict the weightage using the neural network
    X = np.array(data)[:, 1:]  # Exclude the ID column
    weighted_values = model.predict(X)
    return weighted_values

# Function to determine if the weighted value is malicious
def is_malicious(weighted_value):
    if weighted_value > 5:
        return True
    else:
        return False

# Main monitoring loop
def monitor_system():
    data_to_write = []
    for i in range(2):
        # Sleep for 1 minute
        time.sleep(60*1)

        # Get the current CPU, Memory, and Disk usage
        cpu_usage = psutil.cpu_percent(interval=1)
        memory_usage = psutil.virtual_memory().percent
        disk_usage = psutil.disk_usage('/').percent

        # Calculate IO usage and convert to megabytes
        io_counters = psutil.disk_io_counters()
        io_usage = (io_counters.read_bytes + io_counters.write_bytes) / (1024 * 1024)  # Convert to MB
        io_usage = round(io_usage, 3)  # Format to 3 decimal places

        # Insert the collected data into the database
        insert_usage(cpu_usage, memory_usage, disk_usage, io_usage)

        # Append the data to write to CSV
        data_to_write.append([i+1, cpu_usage, memory_usage, disk_usage, io_usage])

    # Display the collected data in a tabular format
    display_usage()
    print("Monitoring complete.")

    # Write the data to a CSV file
    write_to_csv(data_to_write, "usage_data.csv")

    # Calculate weightage using the neural network model
    weighted_values = calculate_weightage(data_to_write)
    print("Weighted values:", weighted_values)

    # Check if the weighted value is malicious
    for weighted_value in weighted_values:
        if is_malicious(weighted_value):
            print("Malicious behavior not detected!")
        else:
            print("Malicious behavior detected.")

if __name__ == "__main__":
    try:
        monitor_system()
    finally:
        # Ensure the cursor and connection are properly closed
        mycursor.close()
        mydb.close()
        print("Database connection closed.")
```

```python
import time
import psutil
import matplotlib.pyplot as plt
import csv
import pandas as pd


# Initialize lists to store data
time_points = []
network_traffic = []

# Function to update data
def update_data(duration):
    for i in range(duration):
        # Get current time
        time_point = time.time()  # Time in seconds
        time_points.append(time_point)

        # Get network traffic
        bytes_sent = psutil.net_io_counters().bytes_sent
        bytes_recv = psutil.net_io_counters().bytes_recv
        total_traffic = (bytes_sent + bytes_recv) / (1024 * 1024)  # Convert to MB
        network_traffic.append(total_traffic)

        # Print network traffic
        print(f"Network Traffic (MB) at {i+1} second(s): {total_traffic}")

        # Write nodes to CSV
        with open('nodes.csv', 'w', newline='') as csvfile:
            fieldnames = ['Time (Seconds)', 'Network (MB)']
            writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
            writer.writeheader()
            row_data = [{'Time (Seconds)': i+1, 'Network (MB)': total_traffic}]
            writer.writerows(row_data)
            df = pd.DataFrame(row_data)

        # Wait for 1 second before updating again
        time.sleep(1)
    return df

# Set the duration for which you want to capture network traffic (in seconds)
duration = 5

# Update data for the specified duration
#update_data(duration)
df = update_data(duration)

#Plot network traffic after the specified duration
plt.plot(time_points, network_traffic, label='Network Traffic (MB)')
plt.xlabel('Time (s)')
plt.ylabel('Network Traffic (MB)')
plt.title('Network Traffic Monitoring')
plt.legend(loc='upper left')
plt.grid(True)
plt.show()

df.to_csv("nodes.csv", index=False)


```
